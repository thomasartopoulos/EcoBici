#### Vemos la cantidad de nulos

```python
def missing_values_table(df_total):
        mis_val = df_total_genero.isnull().sum()
        mis_val_percent = 100 * df_total.isnull().sum() / len(df_total)
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        print ("El DataFrame seleccionado tiene " + str(df_total.shape[1]) + " columnas.\n"      
            "There are " + str(mis_val_table_ren_columns.shape[0]) +
              " columnas que tienen datos faltantes.")
        return mis_val_table_ren_columns
```

```python
missing_values_table(df_total_genero)
```

#### Dropeamos los nulos

```python
df_total_genero = df_total_genero.dropna()
missing_values_table(df_total_genero)
```

### Retocamos

```python
df_total_genero["fecha_origen_recorrido"] = pd.to_datetime(df_total_genero['fecha_origen_recorrido'])
df_total_genero["fecha_destino_recorrido"] = pd.to_datetime(df_total_genero['fecha_destino_recorrido'])
```

```python
df_total_genero[["id_estacion_origen", "id_estacion_destino"]] = df_total_genero[["id_estacion_origen", "id_estacion_destino"]].astype(float).round()
```

```python
df_total_genero[["id_estacion_origen", "id_estacion_destino"]] = df_total_genero[["id_estacion_origen", "id_estacion_destino"]].astype(int)
```

```python
df_total_genero[["id_estacion_origen", "id_estacion_destino"]].dtypes
```

```python
# df_total_genero[["id_estacion_origen","id_estacion_destino"]]   = df_total_genero[["id_estacion_origen","id_estacion_destino"]].astype(int)
```

```python
# Se crea la categoría "día" de la semana iterando sobre los siete días (4/3/2019 es lunes)
days = [(dt.datetime(2019, 3, 4) + dt.timedelta(days=x)).strftime('%a') for x in range(0, 7)]
day_type = CategoricalDtype(categories=days, ordered=True)

# Se crea la categoría "mes" iterando sobre los meses
months = [dt.datetime(2019, x, 1).strftime('%B') for x in range(1, 13)]
month_type = CategoricalDtype(categories=months, ordered=True)
```

```python
# Se limpia en nombre de columna por caso de uso
#df.columns = [' '.join(x.replace('trip_', '').replace('_seconds', '').split('_')).title() for x in df.columns]

# df_total["fecha_origen_recorrido"] = pd.to_datetime(df_total["fecha_origen_recorrido"])

df_total_genero['Date'] = df_total_genero['fecha_origen_recorrido'].apply(lambda x: x.strftime('%Y-%m-%d'))
df_total_genero['Quarter'] = df_total_genero['fecha_origen_recorrido'].apply(lambda x: int((int(x.strftime('%m')) - 1) / 3) + 1)
df_total_genero['Month'] = df_total_genero['fecha_origen_recorrido'].apply(lambda x: x.strftime('%B')).astype(month_type)
df_total_genero['Day of Week'] = df_total_genero['fecha_origen_recorrido'].apply(lambda x: x.strftime('%a')).astype(day_type)
df_total_genero['Hour'] = df_total_genero['fecha_origen_recorrido'].apply(lambda x: x.strftime('%H'))

# Se crea la columna "Route ID" con el formato "[ID Estacion de Origen]-[ID Estacion de destino]" 
df_total_genero['Route Id'] = df_total_genero.apply(lambda x: '{0}-{1}'.format(int(x['id_estacion_origen']), int(x["id_estacion_destino"])), axis=1)
```

```python
df_total_genero.to_csv("df_total_genero_normalizado.csv")
```
