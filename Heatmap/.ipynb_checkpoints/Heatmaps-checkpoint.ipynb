```python
import pandas as pd
import numpy as np
from datetime import timedelta
import matplotlib.pyplot as plt
import requests
import seaborn as sns
import json
import os
import datetime as dt
import re
import folium
from folium.plugins import HeatMap
import folium.plugins as plugins


from geopy import distance
from pandas.api.types import CategoricalDtype
```

# Helper Functions

```python
m = folium.Map([-34.62, -58.4], zoom_start=12)

def crear_heatmap(df, labels): # labes = lat, lon y variable a plotear
    data_heat = df[labels]
    m = folium.Map([-34.62, -58.4], zoom_start=12)
    for i in range(len(df)):
        folium.Circle(
                location=[df.iloc[i][labels[0]], df.iloc[i][labels[1]]],
#                 print (location),
                radius=10,
    #             color = color,
                opacity=0.8,
    #             popup = ('Precio: ' + str(df.iloc[i].price_usd_per_m2).capitalize() + '<br>'
    #                      'Tipo de casa: ' + str(df.iloc[i].property_type) + '<br>'

                        ).add_to(m)
    
    m.add_child(HeatMap(data_heat, 
                        radius=8.4, 
                        min_opacity=1,
                        max_val =2,
                        blur=15,
                        gradient ={0.25: "blue", 
                                  0.50: "green",                              
                                  0.75: "yellow", 
                                  1: "red"                
                       }))
    folium.LayerControl().add_to(m)
    return m
```

# Armado de datasets

```python
df = pd.read_csv("recorridos bicicleta/reducido_genero.csv", index_col="Unnamed: 0.1")
hombres = df.loc[df.genero_usuario=="M"]
mujeres = df.loc[df.genero_usuario=="F"]
df = df.drop(columns="Unnamed: 0" )
df.sort_index(inplace=True)
```

```python
hombres = df.loc[df.genero_usuario == "M"]
mujeres = df.loc[df.genero_usuario == "F"]
```

### Convirtiendo a variables datetime

```python
df["fecha_origen_recorrido"] = pd.to_datetime(df['fecha_origen_recorrido'])
df["fecha_destino_recorrido"] = pd.to_datetime(df['fecha_destino_recorrido'])
df["duracion_recorrido"]= (df["fecha_destino_recorrido"]-df["fecha_origen_recorrido"])
# df["duracion_recorrido"]
```

# Distribuci√≥n de la duracion de los viajes por genero

### Preprocesamiento por genero

```python
# # # Para hombres

hombres_populares_origen = hombres.groupby(by="id_estacion_origen").count()["periodo"].sort_values(ascending=False)
# hombres_populares
hombres_populares_destino = hombres.groupby(by="id_estacion_destino").count()["periodo"].sort_values(ascending=False)

hombres_populares = pd.DataFrame()
hombres_populares["origen"] = hombres_populares_origen
hombres_populares["destino"]=hombres_populares_destino
# hombres_populares.rename(columns={"periodo":"destino"})
hombres_populares["total"] =  hombres_populares["origen"] + hombres_populares["destino"]
hombres_populares["genero"] = "M"

hombres_top5_estaciones_origen = hombres_populares["origen"].sort_values().tail(5).index.tolist()
hombres_top5_estaciones_destino = hombres_populares["destino"].sort_values().tail(5).index.tolist()
```

```python
# # # Para mujeres

mujeres_populares_origen = mujeres.groupby(by="id_estacion_origen").count()["periodo"].sort_values(ascending=False)
# hombres_populares
mujeres_populares_destino = mujeres.groupby(by="id_estacion_destino").count()["periodo"].sort_values(ascending=False)

mujeres_populares = pd.DataFrame()
mujeres_populares["origen"] = mujeres_populares_origen
mujeres_populares["destino"]= mujeres_populares_destino
# hombres_populares.rename(columns={"periodo":"destino"})
mujeres_populares["total"] =  mujeres_populares["origen"] + mujeres_populares["destino"]
mujeres_populares["genero"] = "F"

mujeres_top5_estaciones_origen = mujeres_populares["origen"].sort_values().tail(5).index.tolist()
mujeres_top5_estaciones_destino = mujeres_populares["destino"].sort_values().tail(5).index.tolist()
```

```python
resultados_genero = pd.DataFrame({"cantidad_viajes": [hombres.shape[0], 
                                                      mujeres.shape[0]],
                                  "duracion_promedio":[hombres.duracion_recorrido.mean(),
                                                      mujeres.duracion_recorrido.mean()],
                                  "estaciones_origen_populares" : [hombres_top5_estaciones_origen,
                                                                   mujeres_top5_estaciones_origen],
                                  "estaciones_destino_populares" : [hombres_top5_estaciones_destino,
                                                                   mujeres_top5_estaciones_destino]}
                                )
resultados_genero = resultados_genero.rename(index={0:"Hombres",
                                                   1:"Mujeres"})
resultados_genero
```

```python
# fig, (ax1, ax2) = plt.subplots(2, 2, figsize=(15, 15))

plt.figure(figsize=(8, 5))
sns.distplot(hombres["duracion_recorrido"])
sns.distplot(mujeres["duracion_recorrido"])
```

```python
# Extract the unique routes from the data with their start and end coordinates
maps = df[['Route Id', 'id_estacion_origen', 'lat_estacion_origen', 'long_estacion_origen', 'id_estacion_destino', 'lat_estacion_destino', 'long_estacion_destino']].drop_duplicates()
maps['Distance'] = maps.apply(lambda x: distance.distance((x['lat_estacion_origen'], x['long_estacion_origen']), (x["lat_estacion_destino"], x['long_estacion_destino'])).km * 1000, axis=1)

# Merge the distance calculation with the main DataFrame
df = df.merge(maps[['Route Id', 'Distance']], how='left', on='Route Id')
df = df[[x for x in df.columns if not 'From' in x and not 'To' in x and x != 'Route Id']]
```

```python
fig, (ax1, ax2) = plt.subplots(2, 2, figsize=(15, 15))
for i, col in enumerate(['duracion_recorrido', 'Distance']):
    # Graph the distribution plots (seaborn doesn't allow default separation for distplot)
    for j, user in enumerate(df['genero_usuario'].unique()):
        sns.distplot(df[df['genero_usuario']==user][col], hist=False, ax=ax2[i])
    
    # Graph the boxplots
    sns.boxplot(x=col, y='genero_usuario', data=df, ax=ax1[i])
    ax1[i].set_xlabel('')
```

```python
df_conid = df.copy()
df_conid.dropna(subset=['id_usuario'], how='all', inplace=True)
```

```python
# Create the day of week Category by iterrating over 7 days (March 4th, 2019 is a Monday)
days = [(dt.datetime(2019, 3, 4) + dt.timedelta(days=x)).strftime('%a') for x in range(0, 7)]
day_type = CategoricalDtype(categories=days, ordered=True)

# Create the month Category by iterrating over the months
months = [dt.datetime(2019, x, 1).strftime('%B') for x in range(1, 13)]
month_type = CategoricalDtype(categories=months, ordered=True)
```

```python
df_conid['Date'] = df_conid['fecha_origen_recorrido'].apply(lambda x: x.date)
```

```python
df_conid['Date'] = df_conid['fecha_origen_recorrido'].apply(lambda x: x.strftime('%Y-%m-%d'))
df_conid['Quarter'] = df_conid['fecha_origen_recorrido'].apply(lambda x: int((int(x.strftime('%m')) - 1) / 3) + 1)
df_conid['Month'] = df_conid['fecha_origen_recorrido'].apply(lambda x: x.strftime('%B')).astype(month_type)
df_conid['Day of Week'] = df_conid['fecha_origen_recorrido'].apply(lambda x: x.strftime('%a')).astype(day_type)
df_conid['Hour'] = df_conid['fecha_origen_recorrido'].apply(lambda x: x.strftime('%H'))
```

```python
# Aggregate the ridership by date fields and user type and count the number of unique trips
ridership = df_conid.groupby(['Date', 'Quarter', 'Month', 'Day of Week', 'genero_usuario'])['id_usuario']\
              .nunique().reset_index().sort_values('Date', ascending=True)
```

```python
p = sns.relplot(x='Date', y='id_usuario', hue='genero_usuario', data=ridership, kind='line', height=9, aspect=16/9.)
p.ax.set_ylabel('Total Daily Trips')

# Limit the number of labels shown on the x-axis to only show the first day of each month
p.ax.set_xticklabels([x.set_text('') if not x.get_text().endswith('-01') else x for x in p.ax.get_xticklabels()], rotation=45)
p.show()
```

# Hetmaps con tiempo

### Estaciones de origen

```python
# # # # # Origen

estaciones_origen = df[['id_estacion_origen', 'nombre_estacion_origen', 'long_estacion_origen',
       'lat_estacion_origen', 'domicilio_estacion_origen']]
estaciones_origen = estaciones_origen.drop_duplicates(subset=("id_estacion_origen")).sort_values("id_estacion_origen").reset_index()
estaciones_origen=estaciones_origen.set_index("id_estacion_origen")#.drop(columns="index")

df["hour"] = df.fecha_origen_recorrido.dt.hour
groupby_viajes_hora_origen = df.groupby(["id_estacion_origen", "hour"])["periodo"].count().unstack(1).replace(np.nan, 0)
groupby_viajes_hora_origen[['lat_estacion_origen', 'long_estacion_origen']] = estaciones_origen [['lat_estacion_origen', 'long_estacion_origen']]
# groupby_viajes_hora_origen
```

```python
### Cada fila es una hora, y cada columna es una estacion
heatmap_origen = pd.DataFrame()
for hora in range (0, 24):
    heatmap_origen[hora] = groupby_viajes_hora_origen.apply(lambda x: [x["lat_estacion_origen"], 
                                                     x["long_estacion_origen"], 
                                                     int(x[hora])/100], axis=1) # 
heatmap_origen = heatmap_origen.transpose()
# heatmap_origen
```

```python
from datetime import datetime, timedelta
sr = pd.Series(pd.date_range('2020-7-12 1:00', 
                       periods = 24, freq = 'H')) 

array_origen = heatmap_origen.to_numpy().tolist()

m = folium.Map([-34.62, -58.4], zoom_start=12)

hm = plugins.HeatMapWithTime(array_origen, 
#                              index=sr, # Que lista de horarios toma para avanzar
                             auto_play=True, 
                             max_opacity=0.3)

hm.add_to(m)

m.save("heatmaps/heatmap_origen.html")
```
